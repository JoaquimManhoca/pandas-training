```python
ufo.head()
```

### Second method

```python
ufo.columns
```

```python
columns_list = ['city', 'colors_reported', 'shape_reported', 'state', 'time']
```

```python
ufo.head()
```

### Third method

```python
columns_list2 = ['City', 'Colors_reported', 'Shape_reported', 'Ttate', 'Time']
ufo2 = pd.read_csv('data/ufo.csv', names=columns_list2, header=0)
```

# How to Drop columns in pandas DataFrame

```python
ufo3 = pd.read_csv('data/ufo.csv')
ufo3.head()
```

### First method Drop one simgle column

```python
ufo3.drop('Colors Reported', axis=1, inplace=True)
```

```python
ufo3.head()
```

### Second method Drop two or more columns

```python
ufo3.drop(['City', 'State'], axis=1, inplace=True)
```

```python
ufo3.head()
```

### Third method Drop one row in DataFrame

```python
ufo3.drop(0, axis=0, inplace=True)
```

```python
ufo3.head()
```

### Fourth method Drop many Rows in DataFrame

```python
ufo3.drop([1,3,5], axis=0, inplace=True)
```

```python
ufo3.head()
```
