# How to Drop columns in pandas DataFrame

```python
ufo3 = pd.read_csv('data/ufo.csv')
ufo3.head()
```

### First method Drop one simgle column

```python
ufo3.drop('Colors Reported', axis=1, inplace=True)
```

```python
ufo3.head()
```

### Second method Drop two or more columns

```python
ufo3.drop(['City', 'State'], axis=1, inplace=True)
```

```python
ufo3.head()
```

### Third method Drop one row in DataFrame

```python
ufo3.drop(0, axis=0, inplace=True)
```

```python
ufo3.head()
```

### Fourth method Drop many Rows in DataFrame

```python
ufo3.drop([1,3,5], axis=0, inplace=True)
```

```python
ufo3.head()
```

 # How do I filter rows of a pandas DataFrame by columns values

```python
movies = pd.read_csv('data/imdb_1000.csv')
movies.head()
```

```python
is_long = movies['duration'] >= 200
```

```python
movies[is_long]
```

```python
movies[movies['duration'] >= 200]
```

### Using two condition

```python
movies[(movies['duration'] >= 200) & (movies['genre'] == 'Action')]
```

### Using more then one condition

```python
movies[movies['genre'].isin(['Action', 'Drama', 'Crime'])]
```

```python
movies[(movies['genre'].isin(['Action', 'Drama', 'Crime'])) & (movies['star_rating'] >= 8)]
```

# When should I use a groupby in pandas

```python
drinks = pd.read_csv('data/drinks.csv')
```

```python
drinks.columns
```

```python
drinks.groupby('continent').beer_servings.mean()
```

```python
drinks[drinks['continent'] == 'Africa'].beer_servings.mean()
```

```python
drinks.groupby('continent').beer_servings.agg(['count', 'min', 'max', 'mean'])
```

```python
drinks.groupby('continent').mean().plot(kind='bar')
```
